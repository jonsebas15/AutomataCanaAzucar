**Implementación del autómata con automata-lib**:

```python
def procesoAzucar():
    from automata.fa.nfa import NFA

    d = NFA(
        states = { 'q1', 'q2', 'q3', 'q4', 'q5', 'q6', 'q7','q8','q9','q10','q11','q12','q13','q14','q15','q16','q17','q18','q19','q20','q21','q22','q23'},
        input_symbols={'c','b','e','j','m','n','o','s','z','i','l'},
        transitions ={
            'q1':{'c':{'q2'}},
            'q2':{'j':{'q5'},
                  'b':{'q3'}},
            'q3':{'e':{'q4'}},
            'q5':{'s':{'q6'},
                  'j':{'q8'}},
            'q6':{'n':{'q7'}},

            'q8':{'m':{'q9'}},
            'q9':{'o':{'q10'}},
            'q10':{'z':{'q13'},
                   'o':{'q11'}},
            'q11':{'o':{'q12'}},
            'q12':{'z':{'q15'},
                   'o':{'q17'}},
            'q13':{'z':{'q14'}},

            'q15':{'z':{'q16'}},

            'q17':{'o':{'q18'}},
            'q18':{'i':{'q19'}},
            'q19':{'i':{'q20','q21'}},

            'q21':{'l':{'q22'}},
            'q22':{'l':{'q23'}}


},
        initial_state='q1',
        final_states={'q4','q7','q14','q16','q20','q23'}
    )
    return d
```


|PROCESOS ACEPTADOS $\hspace{2.0cm}$ |  PROCESOS RECHAZADOS $\hspace{2.0cm}$|

|$cjsn         $                 |$cb\hspace{0.1cm}                $   |
|$cjjmozz       $                      |$cjjmoz\hspace{0.1cm} $                |
|$cjjmooozz  $                            |$cjjmoooooi\hspace{0.1cm} $              |
|$cjjmoooooii   $                                       |$cjs\hspace{0.1cm} $
 |$cjjmoooooiill$         | $cjjmooo\hspace{0.1cm}$   |

```python
def automata():
    d = procesoAzucar()
    valid_words = ['cbe',
                   'cjsn',
                   'cjjmozz',
                   'cjjmooozz',
                   'cjjmoooooii',
                   'cjjmoooooiill',
                   ]

    invalid_words = ['cjjmooooiil',
                     'cb',
                     'cjjmoz',
                     'cjjmoooooi',
                     'cjs',
                     'cjjmooo']

    for i in valid_words:
        if d.accepts_input(i) != True:
            return False
    for j in invalid_words:
        if d.accepts_input(j) != False:
            return False
    return True

if automata():
  print('El proceso es correcto ')
else:
  print('MAL! El diseño es incorrecto. Sigue intentando')
```

```python
def palabra(val):
  if(val=='cbe'):
    return print("Se dio energia a los generadores satifactoriamente")
  if(val=='cjsn'):
    return print("Se logró recolectar abono")
  if(val=='cjjmozz'):
    return print("Se ha producido azucar blanca")
  if(val=='cjjmooozz'):
    return print("Se ha producido azucar fina")
  if(val=='cjjmoooooii'):
    return print("Se ha producido miel")
  if(val=='cjjmoooooiill'):
    return print("Se ha producido alcohol")
  return print("Palabra no aceptada")
```
